## 8️⃣ RÉSULTATS — Ce qu'On a Appris

### Explorations des Données (EDA)

#### Figure 1: Distributions des 12 Canaux 📊
```
Les 12 canaux d'entrée:
1. DEM (altitude)           → -500 à +4000 m
2. Slope (pente)            → 0 à 90°
3. wind_speed               → 0 à 20 m/s
4. wind_direction           → 0 à 360°
5. min_temperature          → -40 à +50°C
6. max_temperature          → -30 à +60°C
7. humidity                 → 0 à 100%
8. precipitation            → 0 à 500 mm
9. NDVI (densité végétation)→ -1 à +1
10. Fire danger index       → Low à High
11. Previous fire mask      → 0 ou 1 (feu hier?)
12. Population              → Log scale
```

**Observations clés**:
- Certains canaux très "skewed" (asymétriques)
- Besoin d'une bonne normalisation
- Corrélations intéressantes entre variables

#### Figure 2: Comparaison Feu vs Pas de Feu 🔥❄️

Quand il Y A du feu demain, les pixels sont généralement:
- ✅ Plus proches d'un feu d'hier (previous_fire)
- ✅ Vitesse du vent plus élevée
- ✅ Humidité plus basse
- ✅ Végétation plus dense

```
Distribution avec feu:     [████████░░░░░░░] ← Plus à droite
Distribution sans feu:     [░░░░░████████░░░] ← Plus à gauche
```

#### Figure 3: Déséquilibre de Classe ⚠️

Problème majeur découvert:
```
Total pixels: ~60 millions
Feu:          ~500,000 (0.8%)
Sans feu:     ~59,500,000 (99.2%)

                      MASSIF déséquilibre!
            Un modèle "paresseux" pourrait dire
            JAMAIS "feu" et avoir 99% accuracy!
```

**Solution appliquée**: Weighted Loss Function
- Pénaliser plus les "faux négatifs" (rater les feux)
- Récompenser les "vrais positifs" (trouver les feux)

### Courbes d'Entraînement 📈

#### Interprétation Simple

```
Loss (Erreur)
    ^
    │     ╱╲ ← Peut monter et descendre
    │    ╱  ╲
    │   ╱    ╲___
    │  ╱         ╲___
    │ ╱              ╲___  ← Finalement descend
    │╱                    ╲___...
    └─────────────────────────→ Epochs
    
- Au début: Loss très élevée (le modèle est nul)
- Milieu: Loss descend (apprend!)
- Fin: Loss plateau (a presque appris tout ce possible)
```

#### Loss vs Val_Loss

```
Training Loss: [████████░░░░░░░] ← Descend bien
Validation Loss: [██████░░░░░░░░░] ← Descend, mais moins bien

Interprétation:
- Si Training Loss << Val Loss = OVERFITTING
```

## 7️⃣ ÉVALUATION — Comment Mesurer la Qualité?

### Le Problème Fondamental 🎯

Pour prédire le feu, chaque pixel peut être:
1. **Feu demain** (classe positive)
2. **Pas de feu demain** (classe négative)

Le modèle fait 4 types de réponses possibles:

|  | Feu Réel | Pas de feu Réel |
|---|----------|-----------------|
| **Modèle dit "FEU"** | ✅ TP (Vrai Positif) | ❌ FP (Faux Positif) |
| **Modèle dit "NON"** | ❌ FN (Faux Négatif) | ✅ TN (Vrai Négatif) |

### Les Métriques Expliquées 📊

#### 1. **Accuracy (Exactitude)** = Correct sur Tout
```
Accuracy = (TP + TN) / (TP + FP + FN + TN)
         = "Combien de prédictions exactes en total?"

Problème: Si 99% des pixels ne sont pas du feu,
          un modèle qui dit JAMAIS "feu" aurait 99% d'accuracy!
          → Pas bon pour notre cas (on perdrait les vrais feux!)
```

#### 2. **Precision (Précision)** = Quand tu dis "Feu", c'est vrai?
```
Precision = TP / (TP + FP)
          = "Parmi tous les feux que tu as prédits,
             combien étaient vraiment des feux?"

Analyse:
- Precision = 1.0: PARFAIT! Zéro faux alerte
- Precision = 0.5: La moitié de tes alertes sont fausses
  
Usage: Important quand les fausses alertes coûtent cher
       (évacuations inutiles = argent, panique)
```

#### 3. **Recall (Rappel)** = Trouves-tu tous les vrais feux?
```
Recall = TP / (TP + FN)
       = "Parmi tous les vrais feux,
          combien tu en as détecté?"

Analyse:
- Recall = 1.0: PARFAIT! Tu n'as raté aucun feu
- Recall = 0.5: Tu en as manqué la moitié

Usage: Crucial pour les incendies!
       Faux négatif = Feu non détecté = CATASTROPHE
```

#### 4. **F1-Score** = L'Équilibre
```
F1 = 2 × (Precision × Recall) / (Precision + Recall)

"Harmonic mean" de Precision et Recall

Avantage: Pénalise les extrêmes
- Si tout Precision, pas bon
- Si tout Recall, pas bon
- F1 force un équilibre
```

#### 5. **Dice Score** = Comme l'Intersection-Over-Union
```
Dice = 2 × (Area d'Intersection) / (Total des deux zones)

Imaginé comme une comparaison de zones de feu:
Zone prédite: [████░░░░░]
Zone réelle:  [░████░░░░]
Intersection: [░██░░░░░░]

Dice mesure: "C'est combien similaire?"
```

#### 6. **IoU (Intersection Over Union)** = Chevauchement
```
IoU = Intersection / Union

Imaginé:
Zone prédite: [████░░░░░]
Zone réelle:  [░████░░░░]
Intersection: [░██░░░░░░]
Union:        [████████░]

IoU = (2) / (8) = 0.25 = 25%
```

#### 7. **AUC(PR) = Area Under Precision-Recall Curve** = Équilibre Global
```
AUC(PR) mesure: "À quel point le modèle est bon en moyenne?"

Courbe PR:
   Precision ↑
   1.0 |       ╱╲
       |      ╱  ╲
   0.5|     ╱    ╲
       |    ╱      ╲___
     0 |___╱___________╲___
       └─────────────────→ Recall
       Aire sous courbe = AUC(PR)

Interprétation:
- AUC(PR) = 1.0: PARFAIT (confiance maximale au bon seuil)
- AUC(PR) = 0.5: Pas mieux qu'aléatoire
- AUC(PR) = 0.7-0.8: Très bon pour déséquilibre extrême

Avantage: Résume la VRAIE performance sur les deux métriques importantes
          (Precision ET Recall) au-dessus de tous les seuils possibles

Pour notre cas (désquilibre 99:1):
- Modèles faibles: AUC(PR) ≈ 0.30-0.40
- Modèles bons: AUC(PR) ≈ 0.65-0.80
- Modèles excellents: AUC(PR) > 0.80

IMPORTANT: AUC(PR) est plus honnête qu'Accuracy!
           (Accuracy ≈ 99% même pour un modèle nul)
```

## 2️⃣ CONCEPTS FONDAMENTAUX

### Qu'est-ce que le "Deep Learning"?
**Deep Learning** = "Apprentissage Profond"

C'est une technique où on crée un **réseau de neurones artificiels** (inspiré du cerveau humain) qui **apprend à partir d'exemples**.

#### Analogie du Cerveau 🧠
- **Notre cerveau**: Apprend à reconnaître les visages en voyant plein de visages
- **Machine Learning**: Pareil, mais pour un ordinateur avec des données numériques

#### Comment ça marche?
```
Données d'entrée (photos, météo) 
                    ↓
            [Réseau de neurones]  ← Énorme calculatrice
                    ↓
         Prédiction de résultat ← "Demain, le feu sera ici"
```

### Qu'est-ce que la "CNN"? (Convolutional Neural Network)
**CNN** = "Réseau de Neurones Convolutionnel"

C'est un type de réseau spécialisé pour **comprendre les images**.

#### Pourquoi avoir besoin pour les images?
Une image c'est juste une grille de pixels avec des chiffres:
```
[125, 130, 128, ...]
[130, 135, 132, ...]
[128, 132, 130, ...]
```

Une CNN **reconnaît les motifs** (patterns):
- "Ici, il y a une ligne rouge" → Peut-être le feu!
- "Ici, les pixels sont similaires" → Probablement un gaz ou fumée

#### Les 3 Opérations Principales d'une CNN

1. **Convolution** - "Chercher des motifs"
   - Le réseau regarde 3×3 pixels à la fois
   - Compte les patterns (rougeurs, textures)
   - Glisse sur toute l'image

2. **Pooling** - "Résumer"
   - Réduit la taille de l'image
   - Garde les informations importantes
   - Comme faire un résumé d'un texte long

3. **Activation** - "Décider si c'est important"
   - Si la valeur est petite → "Non pertinent, oublie"
   - Si la valeur est grande → "Oui! Garde-le"
